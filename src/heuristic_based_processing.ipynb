```python
import re

def clean_doc(input_path, output_path):
  with open(input_path, 'r') as f:
    lines = f.readlines()

  cleaned_lines = []
  for line in lines:
    line = line.strip()
    li = re.split('\W+', line)

    cntWord, cntNumber = 0, 0
    for token in li:
      if token.isalpha():
        cntWord += 1
      if token.isnumeric():
        cntNumber += 1

    if cntWord >= 5 and cntNumber < 5:
      cleaned_lines.append(line)

  if len(cleaned_lines) > 0:
    clean_text = " ".join(cleaned_lines)

    with open(output_path, 'w') as f:
      f.write(clean_text)
```

```python
import os

# clean all docs in tesseract
new_dir = 'tesseract_clean/'
if not os.path.exists(new_dir):
  os.makedirs(new_dir)

old_dir = 'tesseract/'
for filename in os.listdir(old_dir):
  old_filepath = os.path.join(old_dir, filename)
  new_filepath = os.path.join(new_dir, filename)
  clean_doc(old_filepath, new_filepath)
  print(new_filepath + " done")
```
